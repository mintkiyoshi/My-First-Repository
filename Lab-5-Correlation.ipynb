# Imports and Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#load database
df = pd.read_csv(r'DATASET\current-power.csv',
                    delimiter= ",")

df
```

# DESCRIBE THE CURRENT AND POWER

```python
df.describe()
```

# COVARIANCE OF CURRENT AND POWER

```python
#covariance of current and power

df.cov()
```

# PLOTTING

```python
#scatter plot

x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current(A)')
plt.show()
```

# CORRELATION COEFFICIENT

```python
#correlation coefficient of current and power

df.corr()
```

```python
#regression line

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#scatter plot
plt.figure(figsize=(7,4))
plt.scatter (x,y,
            color = '#0000FF',
            alpha = 0.6,
            label = 'Data Points')

#line plot
plt.plot (x, line,
          color = '#0000FF',
          label = f'Regression line (r= {r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current(A)')
plt.legend()
```

AIR QUALITY

```python
#load dataset

air = pd.read_csv(r'DATASET\air-quality.csv',
                    delimiter=",")
air
```

# NOx Humidity

```python
#regression line

x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#scatter plot
plt.figure(figsize=(7,4))
plt.scatter (x,y,
            color = '#0000FF',
            alpha = 0.2,
            label = 'Data Points')

#line plot
plt.plot (x, line,
          color = "#AE1F1F",
          label = f'Regression line (r= {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```

# NO2 Humidity

```python
#regression line

x = air['NO2']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#scatter plot
plt.figure(figsize=(7,4))
plt.scatter (x,y,
            color = '#0000FF',
            alpha = 0.2,
            label = 'Data Points')

#line plot
plt.plot (x, line,
          color = "#AE1F1F",
          label = f'Regression line (r= {r_value:.3f})')

plt.title('NO2 vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NO2 (ug/cu.m)')
plt.legend()
plt.show()
```
